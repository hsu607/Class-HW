## Problem 1
1> Draw the graph of y =  sin(x) and y = cos(x) as follows with xticks and yticks.




<font color='red'><<<**Finally Answer**>>></font>

````python
from numpy import pi
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
 
ax.set_xlim([-3.5, 3.5])
ax.set_ylim([-1.2, 1.2])
ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_xticklabels(['$-\pi$', '$-\pi/2$', '$0$', '$\pi/2$', '$\pi$'])
ax.set_yticks([-1, 0, 1])
ax.set_yticklabels(['$-1$', '$0$', '$+1$'])
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
x = np.linspace(-pi, pi, num=100)
y1 = np.cos(x)
y2 = np.sin(x)
plt.plot(x,y1, c = "b")
plt.plot(x,y2, c = "r")
plt.show()
````

````python
from numpy import pi
from math import sqrt
ax = plt.figure(figsize = (16,6))
 
ax.set_xlim([-3.5, 3.5])
ax.set_ylim([-1.2, 1.2])
ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_xticklabels(['$-\pi$', '$-\pi/2$', '$0$', '$\pi/2$', '$\pi$'])
ax.set_yticks([-1, 0, 1])
ax.set_yticklabels(['$-1$', '$0$', '$+1$'])

ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

x = np.linspace(-pi, pi, num=100)
y1 = np.cos(x)
y2 = np.sin(x)
plt.plot(x,y1, c = "b")
plt.plot(x,y2, c = "r")
plt.plot([2*pi/3, 2*pi/3], [sqrt(3)/2, 0], linestyle = "--", c = "r")
plt.plot([2*pi/3, 2*pi/3], [-1/2, 0], linestyle = "--", c = "b")
plt.scatter([2*pi/3, 2*pi/3], [sqrt(3)/2, -1/2], c = ["r", "b"])

ax.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$', xy=(2*pi/3, sqrt(3)/2),
            xytext=(2.3, 1.1), arrowprops = dict(arrowstyle = "->"))
ax.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$', xy=(2*pi/3, -1/2),
            xytext=(0.6, -0.8), arrowprops = dict(arrowstyle = "->"))
plt.legend(['cosine', 'sine'])
plt.show()
````

4> (Video Record) Given the following parametrization of a Mobius strip

Hint 1: You first plot points 30*30 points on 3D space using the following
instruction.

a> Make use of the following code to produce meshgrids of (s,t)

b> You can make x, y, and z from the meshgrids of (s,t)

c> Use np.raven to flatten x, y, and z.

Hint 2: You visualize Mobious strip based on the surface triangulations.

Hint 3: Result is similar to the below figure:

[video solution](https://youtu.be/_ehL8uxeExs)

<font color='green'> Explanation starts here</font>



````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.tri import Triangulation

s = np.linspace(0, 2 * np.pi, 30)
t = np.linspace(-0.5, 0.5, 30)
s, t = np.meshgrid(s, t)

r = 1 + t/2 * np.cos(s/2)

x = np.ravel(r * np.cos(s))
y = np.ravel(r * np.sin(s))
z = np.ravel(t/2 * np.sin(s))

tri = Triangulation(np.ravel(t), np.ravel(s))
````

<font color='green'> Explanation ends here</font>


<font color='red'><<<**Finally Answer**>>></font>

````python
ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z, triangles=tri.triangles, cmap='viridis', linewidths=0.001)
ax.set_xlim(-1.3, 1.3)
ax.set_ylim(-1.3, 1.3)
ax.set_zlim(-0.25, 0.25)
ax.set_xticks([-1, -0.5, 0.0, 0.5, 1])
ax.set_yticks([-1, -0.5, 0.0, 0.5, 1])
ax.set_zticks([-0.2, -0.1, 0.0, 0.1, 0.2])
plt.tight_layout()
````
