## Problem 3
(List and loop) Write a Python program to print the even numbers from a given
list.

Sample List : [1, 2, 3, 4, 5, 6, 7, 8, 9]

Expected Result : [2, 4, 6, 8]

<font color='green'> 풀이과정 :</font>

주어진 리스트 안에 있는 짝수만 골라내야 하므로 for문과 if문을 이용하여 i % 2 == 0 을 만족시키는 값만 새로운 리스트 b에 저장해야 한다.

# <font color='red'><<<**Finally Answer**>>></font>

````python
a = [1, 2, 3, 4, 5, 6, 7, 8, 9]
b = []
for i in a:
    if i % 2 == 0:
        b.append(i)
print(b)
````

<font color='green'> 풀이과정 :</font>

첫번째 행에 해당하는 L = [1]을 프린트해두고, for 루프를 이용해서 나머지 n-1행에 해당하는 리스트를 프린트한다. for문 안에서 L의 첫 번째 원소에 0을 추가한 Left라는 리스트와, for문 안에서 L의 마지막 원소에 0을 추가한 Right라는 리스트를 만든다. L을 빈 리스트로 만들어 다음행에 들어갈 원소를 추가하기 위한 리스트를 만든다. 이중 for 문을 사용하여 Left와 Right의 같은 인덱스에 해당하는 원소들을 더한 값을 L에 저장한 후 각 실행마다 L값을 프린트한다. 이때 Left는 L의 첫 번째 원소에 0을 추가했고, Rigth는 L의 마지막 원소에 0을 추가했기 때문에 Left와 Right를 합친 L은 항상 첫 번째 원소와 마지막 원소가 1로 유지된다. 또한 Left와 Right를 합친 값은 L의 연속된 원소들을 두 개씩 더한 것과 같은 값을 의미한다.

# <font color='red'><<<**Finally Answer**>>></font>

````python
n=3
L=[1]
print(L)

for i in range(n-1):
  Left = [0]+L
  Right = L+[0]
  L=[]
  for j in range(len(Left)):
    L.append(Left[j]+Right[j])
  print(L)
````

## Problem 5
(Loop) Write a Python program to find a pair with highest product from a given
array of integers.

Original array: [1, 2, 3, 4, 7, 0, 8, 4]

Maximum product pair is: (7, 8)

Original array: [0, -1, -2, -4, 5, 0, -6]

Maximum product pair is: (-4, -6)

<font color='green'> 풀이과정 :</font>

for문과 pop() 메소드를 이용해 주어진 array의 원소들을 앞에서부터 차례로 하나씩 뽑아내고 이중 for 문을 사용해 나머지 원소들과의 곱을 계산한 것을 temp에 저장하여 현재 최댓값보다 크면 temp값이 현재 최댓값을 대체하는 루프를 작성한다.   

# <font color='red'><<<**Finally Answer**>>></font>

````python
a = [1, 2, 3, 4, 7, 0, 8, 4]
current_max = a[0] * a[1]
for i in range(len(a)):
    x1 = a.pop(0)
    for x2 in a:
        temp = x1 * x2
        if temp > current_max:
            current_max = temp
            pair_max = [x1, x2]
pair_max
````

````python
a = [0, -1, -2, -4, 5, 0, -6]
current_max = a[0] * a[1]
for i in range(len(a)):
    x1 = a.pop(0)
    for x2 in a:
        temp = x1 * x2
        if temp > current_max:
            current_max = temp
            pair_max = [x1, x2]
pair_max
````

## Problem 6
(Algorithm) You are given the following list:

my_num = [3,5,10,7,2,4]

Write Python code to find the maximum value and the index of the maximum
value in “my_num“. In this case, the maximum value is ”10“ and the index of the
maximum value is ”2”. (Assume that there are no ties.)

<font color='green'> 풀이과정 :</font>

리스트의 첫번째 값을 current_max로 설정하고, for문을 이용해 다음 원소들이 current_max보다 크면 해당 값이 current_max로 대체되고, 해당 인덱스가 current_max_index가 되도록 루프를 작성한다. for문을 다 돌고 나면 current_max와 current_max_index를 프린트한다.

# <font color='red'><<<**Finally Answer**>>></font>

````python
my_num = [3,5,10,7,2,4]
current_max = my_num[0]
for i in range(1,len(my_num)):
    if my_num[i] > current_max:
        current_max = my_num[i]
        current_max_index = i
print('maximum vlaue is', current_max, ', and the index of the maximum value is', current_max_index)
````

[Video] (Algorithm) Given a list of integers with duplicate elements in it, write a Python code to generate another list, which contains only the duplicate elements. In simple words, the new list should contain the elements which appear more than one.


[video solution](https://youtu.be/u2RSXDppV9A)

<font color='green'> 풀이과정 :</font>

이중 for문을 이용하여 리스트 내 원소들을 중복없이 2개씩 조합을 선택하여 선택된 두 개의 값이 같으면 S라는 새로운 집합에 원소를 추가한다. list 대신 set()을 이용하여 원소이름이 여러번 저장되지 않도록 한다.

# <font color='red'><<<**Finally Answer**>>></font>

````python
A = ["Tom", "Jerry", "Mike", "Tom", "Mike"]
S = set()
for i in range(len(A)-1):
  for j in range(i+1, len(A)):
    if A[i] == A[j]:
      S.add(A[i])

list(S)
````
