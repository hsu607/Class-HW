## Problem 1
(Function) Write a Python function to check whether a number is perfect or
not.


<font color='green'> Explanation starts here</font>

자기 자신을 제외한 약수를 모두 더하는 코드 작성 

```
mysum = 0
for i in range(1, n):
    if n % i == 0:
        mysum += i
```

mysum이 입력한 숫자와 같으면 perfect number을 출력하고 아니면 Not perfect number을 출력하는 코드 추가

<font color='green'> Explanation ends here</font>


<font color='red'><<<**Finally Answer**>>></font>

````python
def check_perfect_num(n):
    mysum = 0
    for i in range(1, n):
        if n % i == 0:
            mysum += i
    if mysum == n:
        print("perfect number")
    else:
        print("Not perfect number")
check_perfect_num(6)
check_perfect_num(28)
check_perfect_num(496)
check_perfect_num(8128)
````

<font color='green'> Explanation ends here</font>


<font color='red'><<<**Finally Answer**>>></font>

````python
from math import pi
class Circle:
    def __init__(self, a):
        self.a = a
        self.myperi = 2 * self.a * pi
        self.myarea = pi*(self.a**2)

    def perimeter(self):
        return self.myperi

    def area(self):
        return self.myarea

a = 3.0
cir0 = Circle(a)
print(cir0.area(), cir0.perimeter())
````

## Problem 6
(Algorithm: recursion)
Using recursive function, write a Python function to get the sum of each digit in a non-negative integer.

<font color='green'> Explanation starts here</font>

```
def sumDigits(n):
    return (n%10) + sumDigits(n//10)
```
에러 발생

마지막에 n//10 값이 0이 되면 0을 return 하고 끝내줘야 함

```
def sumDigits(n):
    if n == 0:
        return 0
    else:
        return (n%10) + sumDigits(n//10)
```

<font color='green'> Explanation ends here</font>


<font color='red'><<<**Finally Answer**>>></font>

````python
def sumDigits(n):
    if n == 0:
        return 0
    else:
        return (n%10) + sumDigits(n//10)

print(sumDigits(345))
print(sumDigits(45))
````

## Problem 7
 (Algorithm: recursion (Towers of Hanoi)

 Write a recursive Python function which explains how to move n disks from rod A to rod C. For example, your function print the following:




<font color='red'><<<**Finally Answer**>>></font>

````python
def TowerOfHanoi(n,a,b,c):
    if n==1:
        print ("Move disk 1 from rod", a,"to rod",b)
        return
    TowerOfHanoi(n-1, a, c, b)
    print ("Move disk",n,"from rod", a,"to rod", b)
    TowerOfHanoi(n-1, c, b, a)

TowerOfHanoi(4,"A","B","C")
````
